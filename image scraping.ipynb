```python
import os
import selenium
from selenium import webdriver
import time
from PIL import Image
import io
import requests
#from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import ElementClickInterceptedException
```

```python
#lets first connect to the web driver
driver = webdriver.Chrome(r"D:\data science\chromedriver.exe")
url="https://www.amazon.in"
driver.get(url)
driver.find_element_by_xpath('//input[@name="field-keywords"]').send_keys('Saree')
driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]').click()
linkesss=[]
for d in range(1,6):
    time.sleep(2)
    lin=driver.find_elements_by_xpath('//img[@class="s-image"]')
    for i in lin:
        linkesss.append(i.get_attribute("src"))
    driver.find_element_by_xpath('//li[@class="a-last"]').click()
    time.sleep(2)
os.chdir('C:/Users/mansnothot/Desktop/Saree')
baseDir=os.getcwd()
for i, url in enumerate(linkesss):
    file_name = f"{i}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        image = image.resize((240,320))
        file_path = os.path.join(baseDir, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG")
        print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
        print(f"ERROR - COULD NOT SAVE {url} - {e}")
```
